## Librerias

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

## Carga y visualización

```python
muscle_img = cv2.imread('../img/MUSCLE.png')

plt.imshow(muscle_img)
plt.show()
```

## Histograma

```python
# histograma con cv2
hist = cv2.calcHist([muscle_img],[0],None,[256],[0,256])

plt.plot(hist)
plt.show()
```

## Solución

Teniendo en cuenta el histograma de muscle y que las fibras rojas son las que tiene tonos mas oscuros en la imagen, se crea una paleta de colores que convierte a blanco los tonos oscuros hasta un valor humbral. De la imagen anterior se determinó un threshold de 85. Por lo tanto, se crea una paleta de color que para valores menores a 85 mapea a 255 mientras que los demas valores los deja intactos.

```python
# paleta de color identidad
lut_fibras_rojas = np.arange(0, 256, 1, dtype=np.uint8)

# se establece valores 255 hasta el valor humbral
lut_fibras_rojas[:85] = 255

# se aplica la paleta de color
dstImage = cv2.LUT(muscle_img , lut_fibras_rojas )

plt.imshow(dstImage)
plt.show()
```

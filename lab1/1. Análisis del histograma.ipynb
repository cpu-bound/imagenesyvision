```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

# Carga de imagenes

```python
clef_img = cv2.imread('../img/CLEF.png', cv2.IMREAD_COLOR)
muscle_img = cv2.imread('../img/MUSCLE.png',cv2.IMREAD_COLOR)
```

# Visualizacion de imagenes

```python
# imshow con matplotlib
plt.imshow(clef_img)
plt.show()
```

```python
# display con PIL
from PIL import Image

display(Image.fromarray(clef_img))
```

```python
fig, ax = plt.subplots(1,2,figsize=(8,16))

ax[0].imshow(clef_img)
ax[1].imshow(muscle_img)
plt.show()
```

# Histograma

```python
# histograma con cv2
hist = cv2.calcHist([clef_img],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```

```python
hist = cv2.calcHist([muscle_img],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```

```python
# histograma con numpy
def show_histogram(image, values=256):
  histogram, bin_edges = np.histogram(image, bins=values, range=(0, values-1))
  fig, ax = plt.subplots()
  ax.set_title("Grayscale Histogram")
  ax.set_xlabel("grayscale value")
  ax.set_ylabel("pixel count")
  ax.set_xlim([0.0, values-1])

  ax.plot(bin_edges[0:-1], histogram)
  plt.show()
```

```python
show_histogram(clef_img)
```

```python
show_histogram(muscle_img)
```

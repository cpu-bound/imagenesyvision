```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

# Carga y visualizacion

```python
regle_img = cv2.imread('../img/REGLE.png')

plt.imshow(regle_img)
plt.show()
```

# histograma

```python
# histograma con cv2
hist = cv2.calcHist([regle_img],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```

# Negativo

```python
negative = cv2.bitwise_not(regle_img)

plt.imshow(negative)
plt.show()
```

```python
# histograma con cv2
hist = cv2.calcHist([negative],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```

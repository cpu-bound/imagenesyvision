```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

# Carga y visualizacion

```python
rondelle_img = cv2.imread('../img/RONDELLE.png', cv2.IMREAD_REDUCED_COLOR_2)

plt.imshow(rondelle_img)
plt.show()
```

```python
spot_img = cv2.imread('../img/SPOT.png', cv2.IMREAD_REDUCED_COLOR_2)

plt.imshow(spot_img)
plt.show()
```

# Histograma

```python
# histograma con cv2
hist = cv2.calcHist([rondelle_img],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```

```python
# histograma con cv2
hist = cv2.calcHist([spot_img],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```

# operaciones aritmeticas
## adicion

```python
addition = cv2.add(rondelle_img, spot_img)

plt.imshow(addition)
plt.show()
```

# operaciones aritmeticas
## sustraccion

```python
substraction = cv2.subtract(rondelle_img, spot_img)

plt.imshow(substraction)
plt.show()
```

```python
substraction_2 = cv2.subtract(spot_img, rondelle_img)

plt.imshow(substraction_2)
plt.show()
```

```python
# histograma con cv2
hist = cv2.calcHist([addition],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```

```python
# histograma con cv2
hist = cv2.calcHist([substraction],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```

```python
# histograma con cv2
hist = cv2.calcHist([substraction_2],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```
